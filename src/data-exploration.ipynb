```python
import pandas as pd

data = pd.read_csv(r"../data/books.csv")
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

ax = plt.axes()
sns.heatmap(data.isna().T, cbar=False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing Values")
plt.show()
```

```python
import numpy as np

data["missing_description"] = np.where(data["description"].isna(), 1, 0)
data["age_of_book"] = 2025 - data["published_year"]

columns_of_interest = [
    "num_pages",
    "age_of_book",
    "missing_description",
    "average_rating",
]

correlation_matrix = data[columns_of_interest].corr(method="spearman")

sns.set_theme(style="dark")
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt=".2f",
    cbar_kws={"label": "Spearman correlation"},
)

plt.show()
```

```python
data[
    (data["description"].isna())
    | (data["num_pages"].isna())
    | (data["average_rating"].isna())
    | (data["published_year"].isna())
]
```

```python
book_missing = data[
    ~(data["description"].isna())
    & ~(data["num_pages"].isna())
    & ~(data["average_rating"].isna())
    & ~(data["published_year"].isna())
]
book_missing
```

```python
book_missing["categories"].value_counts().reset_index().sort_values(
    "count", ascending=False
)
```

```python
book_missing = book_missing.copy()
book_missing.loc[:, "words_in_description"] = book_missing["description"].str.split().str.len()
book_missing
```

```python
mask = book_missing["words_in_description"].between(5, 10)
book_missing.loc[mask, "description"]
```

```python
mask = book_missing["words_in_description"].between(15, 24)
book_missing.loc[mask, "description"]
```

```python
mask = book_missing["words_in_description"].between(25, 34)
book_missing.loc[mask, "description"]
```

```python
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25]
book_missing_25_words
```

```python
book_missing_25_words = book_missing_25_words.copy()

book_missing_25_words.loc[:, "title_and_subtitle"] = np.where(
    book_missing_25_words["subtitle"].isna(),
    book_missing_25_words["title"],
    book_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1),
)
book_missing_25_words
```

```python
book_missing_25_words["description"] = book_missing_25_words["description"].str.replace(
    '"', "", regex=False
)
book_missing_25_words.description
```

```python
book_missing_25_words.drop(
    columns=["subtitle", "missing_description", "age_of_book", "words_in_description"],
    axis=1,
).to_csv(r"./data/books_cleaned.csv", index=False)
```
