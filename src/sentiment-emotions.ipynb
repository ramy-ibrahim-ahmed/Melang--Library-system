```python
import pandas as pd

data = pd.read_csv(r"../data/books_cleaned.csv")
description_1 = data["description"][0]
```

```python
from transformers import pipeline

classifier = pipeline(
    "text-classification",
    model="j-hartmann/emotion-english-distilroberta-base",
)

classifier(description_1)
```

```python
classifier(description_1.split("."))
```

```python
def get_emption(classifier, sentence_of_sentences):
    sentences = classifier(sentence_of_sentences.split("."))
    hashmap = dict()
    for sentence in sentences:
        if sentence["label"] not in hashmap:
            hashmap[sentence["label"]] = sentence["score"]
        else:
            hashmap[sentence["label"]] += sentence["score"]

    maximun = 0
    label = None
    for key, value in hashmap.items():
        if value > maximun and key != "neutral":
            maximun = value
            label = key

    if not label:
        label = "neutral"

    return label


print(get_emption(classifier, description_1))
```

```python
data["emotion"] = data["description"].apply(lambda s: get_emption(classifier, s))
data
```

```python
data.to_csv(r"./data/books_emotions.csv", index=False)
```
